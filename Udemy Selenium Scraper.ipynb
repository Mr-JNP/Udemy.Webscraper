```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
```

```python
chrome_options = Options()
#chrome_options.add_argument("--disable-extensions")
#chrome_options.add_argument("--disable-gpu")
#chrome_options.add_argument("--no-sandbox") # linux only
chrome_options.add_argument("--headless")

driver = webdriver.Chrome(options=chrome_options)

url = "https://www.udemy.com/courses/it-and-software/?p={page}"
big = []
columns = ['name', 'description', 'instructors', 'learners', 'rating'] 

try:
    for i in range (1, 100):
        driver.get(url.format(page=i))

        course_container = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "course-directory--container--5ZPhr")))
        courses = course_container.find_elements_by_class_name('course-card--has-price-text--1Ikr0')

        for course in courses:
            name = course.find_element_by_class_name('course-card--course-title--2f7tE').text
            description = course.find_element_by_class_name('course-card--course-headline--yIrRk').text
            instructors = course.find_element_by_class_name('course-card--instructor-list--lIA4f').text
            learners = course.find_element_by_class_name('course-card--reviews-text--12UpL').text
            rating = course.find_element_by_class_name('star-rating--rating-number--3lVe8').text

            a = [name, description, instructors, learners, rating]
            big.append(a)

        time.sleep(5.0)
finally:
    df = pd.DataFrame(big, columns=columns)
    driver.quit()
```
